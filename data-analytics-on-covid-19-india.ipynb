```python
# Storing total cases (as of 2021-08-11) state wise

stateCases = df[df["Date"] == "2021-08-11"]
```

```python
stateCases.head()
```

```python
# Sorting States in decreasing order based on number of cases

stateCases.sort_values(by="Confirmed",ascending = False)
```

```python
Top_10_States = stateCases.sort_values(by="Confirmed",ascending = False).head(10)
```

```python
# Based on number of COVID-19 Cases
Top_10_States
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.barplot(x = "State/UnionTerritory",y = "Confirmed",data = Top_10_States.head())
```

```python
# Correlation between each other
df.corr()
```

```python
f,ax = plt.subplots(figsize = (10,9)) 
sns.heatmap(df.corr(), annot=True,fmt='.3f',ax=ax,vmin=-1, vmax=1, center= 0, cmap= 'coolwarm',linewidths=4, linecolor='white')
plt.show()
```

#### Total Deaths

```python
plt.figure(figsize=(18,10))
plt.bar(df["State/UnionTerritory"],df["Deaths"])
plt.title("Total Deaths of Patients",fontsize=15)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Patients",fontsize=14)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.show()
```

#### Total Patients

```python
plt.figure(figsize=(18,10))
plt.bar(df["State/UnionTerritory"],df["Confirmed"])
plt.title("Total Active Cases By State",fontsize=15)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Cases",fontsize=14)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.show()
```

#### Final Overview : COVID-19 India

```python
Deaths=df["Deaths"].sum()
Cured=df["Cured"].sum()
Total_Cases=df["Confirmed"].sum()
Active = Total_Cases - Cured 
```

```python
slices=[Active,Deaths,Cured,Total_Cases]
labels=["Active","Deaths","Cured","Total_Cases"]
plt.title("COVID 19 : India OverView")
colors=["#fc9303","#a22502","#AFB4FF","#319DA0"]
plt.pie(slices,colors=colors,autopct='%1.1f%%',labels=labels,shadow=True)
plt.show()
```
